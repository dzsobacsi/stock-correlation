```python
import pandas as pd
import plotly.express as px

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('data/close_prices.csv')
df = df.dropna(thresh=10)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df.head()
```

```python
returns = df.pct_change(fill_method=None)
returns = returns.dropna(how='all')
returns.head()
```

```python
returns.to_csv('data/returns.csv')
```

# Appendix

## Transforming the price data to daily returns provides linearly dependent variables with normal distribution

### Plot of the close prices of two selected stocks

```python
fig = px.line(df, x=df.index, y=['AMZN', 'AAPL'], 
              title='Close prices of two selected stocks')
fig.show()
```

### The price data alone does not allow us to calculate correlation

```python
fig = px.scatter(df, x='AMZN', y='AAPL', 
                 title='Scatter plot of the closing prices', 
                 marginal_x='histogram', marginal_y='histogram', 
                 width=1200, height=600)
fig.show()
```

### Let's calculate and plot the daily returns instead

```python
fig = px.line(returns, x=returns.index, y=['AMZN', 'AAPL'],
              title='Daily returns of two selected stocks')
fig.show()
```

### The daily returns are linearly dependent and normally distributed variables, allowing us to use the Pearson correlation coefficient

```python
fig = px.scatter(returns, x='AMZN', y='AAPL', 
                 title='Scatter plot of the daily returns', 
                 marginal_x='histogram', marginal_y='histogram',
                 width=1200, height=600)
fig.show()
```
