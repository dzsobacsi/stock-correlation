```python
import ast
import pandas as pd
import plotly.express as px

pd.set_option('display.max_columns', None)
```

```python
prices = pd.read_csv('data/close_prices.csv')
prices = prices.set_index('Date')
prices = prices.drop(columns=['COIN'])
prices.head()
```

```python
first_row = prices.iloc[0]
scaling_factors = 100 / first_row
rel_prices = prices * scaling_factors
rel_prices.head()
```

```python
best_sets = pd.read_csv('data/best_sets.csv')
best_sets
```

```python
cols = list(ast.literal_eval(best_sets.iloc[0, 0]))
portfolio = rel_prices[cols]
portfolio['Portfolio value'] = portfolio.sum(axis=1)
portfolio.head()
```

```python
portfolio.tail()
```

```python
fig = px.line(portfolio, x=portfolio.index, y='Portfolio value', 
              title='Portfolio peprformance from 2021-01-04 to 2024-08-08 - Curde Oil, Gold, Meta, Solana',
              width=800, height=500)
fig.update_layout(yaxis_type='log')
fig.show()
```
